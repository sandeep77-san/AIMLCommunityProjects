Before proceeding to step 6, we will learn an important concept which is called label encoding. While navigating through the data, you would have seen that the classification has been done in form of words B and M for benign and malignant cancer cells. But we need this to be converted in number so that this parameter can also be used statistically. 

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_Y=LabelEncoder()
cancer_data.iloc[:,1]=labelencoder_Y.fit_transform(cancer_data.iloc[:,1].values)
# cancer_data.head(cancer_data.shape[0])
```

In the above code, we have applied label encoding on all the rows and one column. #### Step 7: Create a correlation matrix and mention strongly, weakly and negatively correlated quantities. 

We tend to drop features having very strong correlation. Why? Think and try to write the answer. I will write this question's answer in week 3 module. 

```python
#Step 7: Create a correlation matrix and mention strongly, weakly and negatively correlated quantities
cancer_data.corr(method="pearson")
```

```python
# We tend to drop features having very strong correlation. Why?
print('from the correlation matrix an obeservation is that correlation of the same varible is 1. Which also infer that stronger the correlation between the two variables, both variables represent the same behaviour. Any one variable can be removed from the data set as the size of matrix will be reduced and computation burden can be decreased.')
```

```python
# assuming that cutoff to decide weakly (>= -0.1 & <= 0.1) and strongly (>0.9) negatively (< -0.9)
cancer_data_corr = cancer_data.corr(method="pearson")
cancer_data_corr_decision = cancer_data_corr.applymap(str) 
for i in range(cancer_data_corr.shape[0]):
    for j in range(cancer_data_corr.shape[1]):
        if  -0.1 <= cancer_data_corr.iloc[i][j] <= 0.1:
            cancer_data_corr_decision.replace(cancer_data_corr_decision.iloc[i][j], 'weak', inplace = True)
        elif cancer_data_corr.iloc[i][j] <= -0.9:
            cancer_data_corr_decision.replace(cancer_data_corr_decision.iloc[i][j], 'Negative', inplace = True)
        elif cancer_data_corr.iloc[i][j] >= 0.9:
             cancer_data_corr_decision.replace(cancer_data_corr_decision.iloc[i][j], 'Strong', inplace = True)
        else:
            cancer_data_corr_decision.replace(cancer_data_corr_decision.iloc[i][j], 'Normal', inplace = True)
cancer_data_corr_decision
```
